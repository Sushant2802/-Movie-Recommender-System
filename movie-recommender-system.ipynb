```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

# Feature Selection 

```python
movies.info()
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['id', 'title', 'overview', 'genres', 'cast', 'keywords', 'crew']]
movies.head(2)
```

```python
movies.shape
```

# Data Preprocesing 

```python
movies.isnull().sum()
```

```python
movies = movies.dropna()
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0]
```

```python
# convert 'string of list' to 'list' use 'import ast'
import ast

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['genres'].head(3)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords'].head(3)
```

```python
movies['cast'][0]
```

```python
#  Each movie get 1st 3 actors name
import ast

def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
movies.head(3)
```

```python
movies['crew'][0]
```

```python
#  fetch 'director' in movies['crew']
import ast

def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['crew']
```

```python
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview']
```

```python
movies.head(4)
```

```python
movies['overview'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(2)
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(3)
```

```python
new_df = movies[['id', 'title', 'tag']]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head(3)
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

## Reduce variations of a word 

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    li = []
    for i in text.split():
        li.append(ps.stem(i))
    return " ".join(li)
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

# Vectorization (text to number)

```python
 # to calculate similarity score between the 2 movies on 'tag' text

# to convert text -> vector  it's text vectorization 
#    and in that we use simple technique is 'bag of words' 

# to combine all tags 
#     and extract most frequent 5000 words
# then check each tag of movie text in that 5000 words 
#         and create vector basis on tag words frequency
# shape : 5000(movies) * 5000(frequent words)

# we not consider stop-words , because we use this for sentence fomation 
#     and they are not contribute in meanining. like  in, a, are. is, etc.
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english', )
```

```python
# by default countVectorizer return scipy sparx matrix , so we convert into numpy array
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors.shape
```

```python
vectors[0:4]
```

## Most 5000 frequent word

```python
features = cv.get_feature_names_out()

print("length of features : ", len(features))
print('\n')
for i in features:
    print(i)
```

```python
# In high dim euclidean distance is not reliable, 
#    so we prefer to use cosine distance means measuring angles

# to get distance of each movie with anther all movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
def recommend(movie): 
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
